# Mean,Median,Mode

```python
# Repalce the empty cell of particular colunm with mean value -

# He sum of all values divided by number of values

import pandas as pd

df=pd.read_csv("dirtydata.csv")
# print(df)

m=df['Calories'].mean()
df['Calories'].fillna(m,inplace=True)
df.head(30)
```

```python
# Now let us print a Mean Value

import pandas as pd

df=pd.read_csv("dirtydata.csv")
# print(df)

m=df['Calories'].mean()
df['Calories'].fillna(m,inplace=True)
df.head(30)

print("The mean value is :",m)
```

```python
# Replace the empty cell of particular column with median value -

# The value in the middle,after sorting in ascending Order.

Median_Value=df['Calories'].median()
df['Calories'].fillna(m,inplace=True)
df.head(25)

print("The median value is ",Median_Value)
```

```python
# Replace the empty cell of particular column with mode - 
# The value that appears most frequently.

Mode=df['Calories'].mode()
df['Calories'].fillna(Mode,inplace=True)
print("The Mode value is :",Mode)
```
