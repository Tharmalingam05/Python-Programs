# Subplots

```python
# SubPlots (1,3,1) -1 rows ,3 colmns and first plot

import numpy as np
import matplotlib.pyplot as plt

day=[1,2,3,4,5,6,7]
temp1=[50,51,52,48,47,49,46]

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.plot(day,temp1,c='g')

plt.title('Plot 1')

print('-----------------')

temp2=[43,42,40,44,33,35,37]
plt.subplot(1,3,2)
plt.plot(day,temp2)
plt.title('Plot 2')

print('-----------------')

avg=[45,48,48,46,40,42,41]

plt.subplot(1,3,3)
plt.plot(day,temp1,color='r')
plt.title('Plot 3')

plt.show()

```

```python
# Add a title to the entire figure

day=[1,2,3,4,5,6,7]
temp1=[50,51,52,48,47,49,46]

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.plot(day,temp1)
plt.title("Plot 1")

temp2=[43,42,40,44,33,35,37]
plt.subplot(1,3,2)
plt.plot(day,temp2)
plt.title("Plot 2")

avg=[45,48,48,46,40,42,41]
plt.subplot(1,3,3)
plt.plot(day,avg)
plt.title("Plot 3")

plt.suptitle("Weather")
plt.show()

```

# Pie Chart

```python
data=[10,20,30,40]
mylables=['color1','color2','color3','color4']
plt.pie(data,labels=mylables,startangle=90)
```

```python
data = [10,20,30,40]
mylabels= ['color1','color2','color3','color4']
exp_value=[0.2, 0, 0, 0]
mycolor=['magenta', 'cyan','yellow','red']
plt.pie(data,labels=mylabels,explode=exp_value,colors=mycolor)
plt.legend(title = 'colors')
plt.show()
```

```python
data = [10,20,30,40]
mylabels= ['color1','color2','color3','color4']
exp_value=[0.2, 0, 0, 0]
mycolor=['magenta', 'cyan','yellow','red']
plt.pie(data,labels=mylabels,explode=exp_value,shadow = True,colors=mycolor)
plt.legend(title = 'colors')
plt.show()
```
